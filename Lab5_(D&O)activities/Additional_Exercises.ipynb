```python
import pandas as pd
from pathlib import Path
import csv
import json

# Your Info
student_id = "TUPM-25-0900"
base_path = Path.home() / "Documents" / "Activity_5_Files"
base_path.mkdir(parents=True, exist_ok=True)

# Helper to ensure files exist for the analysis
(base_path / f"intro_{student_id}.txt").write_text(f"Welcome Cricylus Garrell Nery (ID: {student_id}) to File Handling!")
(base_path / f"lines_{student_id}.txt").write_text("Line 1: Electronics\nLine 2: Communications\nLine 3: Engineering")
(base_path / f"merged_{student_id}.txt").write_text("This is a merged test file for the complexity analyzer.")
```

```python
# Exercise 1: Complexity Analyzer
results = []
for file in base_path.glob("*.txt"):
    text = file.read_text()
    lines = [line for line in text.splitlines() if line.strip()] # Ignore empty lines
    words = text.split()
    chars = len(text.replace(" ", "").replace("\n", ""))


    wpl = len(words) / len(lines) if lines else 0
    cpw = chars / len(words) if words else 0
    
    results.append({
        "Filename": file.name,
        "Lines": len(lines),
        "Words": len(words),
        "Characters": chars,
        "Words/Line": round(wpl, 2),
        "Chars/Word": round(cpw, 2)
    })

df_complexity = pd.DataFrame(results).sort_values(by="Words/Line", ascending=False)

print("--- Tabular Summary ---")
display(df_complexity)

# Identify Most and Least dense
print(f"\nMost content-dense file: {df_complexity.iloc[0]['Filename']}")
print(f"Least content-dense file: {df_complexity.iloc[-1]['Filename']}")

df_complexity.to_csv(base_path / "content_summary.csv", index=False)
```

```python
# Exercise 2: Word Frequency Analyzer
import re
from collections import Counter

# Define Stopwords
stopwords = {"the", "is", "and", "to", "in", "a", "of", "it", "for", "with", "id"}

all_words = []

for file in base_path.glob("*.txt"):
    text = file.read_text().lower() # Normalize to lowercase
    # Remove punctuation using RegEx
    clean_text = re.sub(r'[^\w\s]', '', text)
    
    # Filter stopwords
    words = [w for w in clean_text.split() if w not in stopwords]
    all_words.extend(words)

# Calculate frequencies
word_counts = Counter(all_words)
freq_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Count']).sort_values(by='Count', ascending=False)

print("--- Most Frequent Meaningful Words ---")
display(freq_df.head(10))

# Export to CSV
freq_df.to_csv(base_path / f"word_frequency_{student_id}.csv", index=False)
```

```python
import shutil
import time
from datetime import datetime
from pathlib import Path

# --- Inputs ---
student_id = "TUPM-25-0900"
student_name = "Cricylus Garrell Nery"

# Define paths
base_dir = Path.home() / "Documents" / "Activity_5_Files"
backup_dir = base_dir / f"backup_{student_id}"
target_file = base_dir / f"intro_{student_id}.txt" # File to back up

# 1. Generate Backup Directory automatically
backup_dir.mkdir(parents=True, exist_ok=True)

if target_file.exists():
    # 2. Generate Timestamp (YYYYMMDD_HHMMSS)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    # 3. Construct unique backup filename
    backup_filename = f"{target_file.stem}_{student_id}_{timestamp}{target_file.suffix}"
    backup_path = backup_dir / backup_filename
    
    # 4. Copy the file
    shutil.copy(target_file, backup_path)
    
    # 5. Update/Create Backup Log
    log_file = base_dir / f"backup_log_{student_id}.txt"
    file_size = backup_path.stat().st_size
    
    log_entry = (
        f"{'-'*30}\n"
        f"Student ID and Name: {student_id} ({student_name})\n"
        f"Original filename: {target_file.name}\n"
        f"Timestamp of backup: {timestamp}\n"
        f"File size in bytes: {file_size}\n"
        f"Backup file path: {backup_path}\n"
    )
    
    with log_file.open("a") as f:
        f.write(log_entry)
        
    # --- Output Confirmation ---
    print(f"Backup completed for {student_id} ({student_name})")
    print(f"File saved as: {backup_filename}")
    print(f"Log updated at: {log_file}")
else:
    print(f"Target file {target_file.name} not found. Please run Snippet 1 first!")
```
